### REGEX

###### ^ a.ka. caret character says that it should start with eg. '^X' means that it should start with 'X'
###### '.' represents any character
###### '*' any number of times; this is a greedy approach i.e. it will return the biggest string that matches the regex
###### '\S' any non-whitespace character
###### '+' one or more times; this is a greedy approach i.e. it will return the biggest string that matches the regex
###### ' ?' this is the non-greedy approach stating that take the smallest string that matches the regex
###### '(' or ')' aren't part of the matching regex but tells from where to start and stop the extraction of the string that matches the regex
###### '[]' represents non-blank character; same as '\S'
###### '\' escape characters; when you want to include special characters which are part of the regex already eg. '\+'
    
    ##### Eg. '^X.*:' this reads as the string that we want to match should START WITH 'X' and then it can have any other characters after it and any number of times and should END WITH ':' .
    #####  Eg. X-DSPAM-Confidence: will return TRUE
    
    ##### Eg. '^X-\S+:' this reads as the string that we want to match should START WITH 'X' FOLLOWED BY '-' and SHOULD HAVE 
    ONE OR MORE NON-WHITESPACE(no spaces), which is represented by '\S+' and SHOULD END WITH ':'
    
    ##### Eg. X-Plane is behind schedule: will return FALSE because there is a space in between the words 
    X-DSPAM-Confidence: will return TRUE because there is no space in between the words and this string satisfies all conditions
    
    ##### Eg. '^X-.+:' this is the greedy approach so it will take the biggest string
    ##### Eg. '^X-.+?:' this is the non-greedy approach so it will take the shortest string; another way to write this is with a
    whitespace Eg. '^X-\S+:'

##### re.search() returns a TRUE/FALSE depending on whether the string matches the regex
##### re.findall() matches the regex and extracts the string that matches with the regex

##### Short Glossary 
    ^	    Matches the beginning of a line
    $	    Matches the end of the line
    .	    Matches any character; wild card expression
    \s	    Matches whitespace
    \S	    Matches any non-whitespace character
    *	    Repeats a character zero or more times
    *?	    Repeats a character zero or more times (non-greedy)
    +	    Repeats a character one or more times
    +?	    Repeats a character one or more times (non-greedy)
[aeiou]	    Matches a single character in the listed set
[^XYZ]	    Matches a single character not in the listed set
[a-z0-9]	The set of characters can include a range
    (	    Indicates where string extraction is to start
    )	    Indicates where string extraction is to end

##### For more information 
https://docs.python.org/3/howto/regex.html

```python
# Extracting strings using regex
import re
x = 'My 2 favourite numbers are 19 and 42.'
y = re.search('[0-9]+',x) # it looks for one or more digits and returns TRUE/FALSE  
z = re.search('[0-9]*',x) # it looks for digits any number of times and returns TRUE/FALSE
v = re.findall('[0-9]+',x) # it looks for one or more digits and returns the string that matches
w = re.findall('[0-9]*',x) # it looks for digits any number of times and returns the string that matches
'''
the difference between using '+' or '*' is that in the case of the former the program will look through the string and will 
return only the strings that match the regex i.e. ['2', '19', '42']. However, the latter will return n-1 positions of the length 
of the string along with the strings that match the regex i.e. ['', '', '', '2', '', '', '', '', '', '', '', '', '', '', '', '',
'', '', '', '', '', '', '', '', '', '', '', '19', '', '', '', '', '', '42', '', '']

''' 
print('y',y,'\n','z',z,'\n','v',v,'\n','w',w)

# extract only email address
u = 'FROM STEPHEN.MARQUARD@UCT.AC.ZA SAT JAN  5 09:14:16 2008'
s= re.findall('\S+@\S+',u) # returns only the email address
t = re.findall('^FROM (\S+@\S+)',u) # returns only the email address but here it matches the 'FROM' word also
print('t',t,'\n','s',s)

# extract the host
r = re.findall('@(\S+)',u)
q = re.findall('@([^ ]*)',u) # same thing but '[]' represents non-blank character; notice the space in between the '[]'
# it tells to start the extract with a single non-blank character till it meets the next whitespace
p = re.findall('^FROM .*@([^ ]*)',u) # this says that i want to start with 'FROM' and then i want a whitespace 
# and then any number of characters until i find '@' and i want to begin extracting all the non-blank characters and 
# i want to stop extracting as soon as i reach a space
print(r,'r','\n',q,'q','\n','p',p)
```

### Data over Networks 

    ###### The way data is send from one computer across networks is  through XML(eXtensible Markup Language) format.
    The difference between XML and HTML is that the former was designed to send data across the network whereas the 
    latter was designed to display data on one's computer. In XML there are no predefined tags such as <p> </p> (paragraph
    tag of HTML). It was invented in Feb 1998. 
    
    For sending data through XML, the origin computer must serialize(sending one character at a time) the data and the 
    destination computer must deserialize(decode the data one character at a time) to be able to view the data. 
    
    The 2 common types of serialization that are commonly used are XML and JSON. 
    
    SCHEMA
    To validate the XML there needs to a be a validation structure(schema). The schema validates all the XML files so that 
    things work. The most common schema is designed by W3C(World Wide Web Cosortium) and the files containing a XML schema
    is stored with the extension of XSD. The XML files are stored with the extension XML. 
    
    Nomenclature
    There are tags, attributes and complex/simple elements. 
    Tags: <person> </person>
    Attributes: <date w="5> </date>
    Complex Elements: 
    <person> 
    <name> </name>
    </person>
    Simple Elements: <name> </name>
    
    
    Dates
    International Organisation for Standardisation(ISO) 8601 is used for the date format. Eg. 2002-05-30
    UTC (Coordinated Universal Time). It's a common time format. No DST is observed. Eg. 2002-05-30T09:30:10Z 
    Z(UTC)
    
    JSON
    Javascript Object Notation:
    This is used for serialisation. The same as XML. The difference is that as the web uses primarily Javascript nowadays 
    it's easier to use a serialisation method which is a derivative of the language that the web uses. 
    It's a language containing dictionaries and nested lists, which translates to ease of use in any programming language as 
    all programming language data structures either support list or dictionaries. In the case of python dictionaries is used
    as a data structure.

```python
# return latitude and longitude of a particular location
import urllib.request, urllib.parse, urllib.error
import xml.etree.ElementTree as ET

serviceurl = 'http://maps.googleapis.com/maps/api/geocode/xml?'

while True:
    address = input('Enter location: ')
    if len(address) < 1: break

    url = serviceurl + urllib.parse.urlencode({'address': address})
    print('Retrieving', url)
    uh = urllib.request.urlopen(url)
    data = uh.read()
    print('Retrieved', len(data), 'characters')
    print(data.decode())
    tree = ET.fromstring(data)

    results = tree.findall('result')
    lat = results[0].find('geometry').find('location').find('lat').text
    lng = results[0].find('geometry').find('location').find('lng').text
    location = results[0].find('formatted_address').text

    print('lat', lat, 'lng', lng)
    print(location)

```

```python
# example of sql in python

import sqlite3 # importing the SQL library to run SQL commands

conn = sqlite3.connect('emaildb.sqlite') # establish connection to the database or create the file
cur = conn.cursor() # this makes a handle through the connection and execute SQL commands through the cursor

cur.execute('DROP TABLE IF EXISTS Counts') # this will drop a database if it does exist and start with a fresh database
# in short it will overwrite the previous database

cur.execute('''
CREATE TABLE Counts (email TEXT, count INTEGER)''') # this creates a table with columns

fname = input('Enter file name: ')
if (len(fname) < 1): fname = 'mbox-short.txt'
fh = open(fname)
for line in fh:
    if not line.startswith('From: '): continue
    pieces = line.split()
    email = pieces[1]
    cur.execute('SELECT count FROM Counts WHERE email = ? ', (email,)) # this puts all the emails in a database 
    # the ? in the code is just a placeholder indicating what value should go in its place; if we have two ? then we can 
    # put two things such as email and name, this ? is followed by a tuple eg. (email,), here we are just going to replace 
    # the ? mark by email.
    row = cur.fetchone() # this fetches the keys or rows; in this case it's all the emails of people;
    if row is None:
        cur.execute('''INSERT INTO Counts (email, count)
                VALUES (?, 1)''', (email,)) # just like a dictionary adds 1 to the count of email and adds the email 
        # similar to updating the dictionary with a get command. here it would translate to counts.get(email,0) +1
    else:
        cur.execute('UPDATE Counts SET count = count + 1 WHERE email = ?',
                    (email,)) # if the email already exists then it just updates the counts of the email
    conn.commit() # writes the the data from memory to the disk

# https://www.sqlite.org/lang_select.html
sqlstr = 'SELECT email, count FROM Counts ORDER BY count DESC LIMIT 10' # this is the code to select the all the emails and 
# return their respective counts; it's stored in the form of a string

for row in cur.execute(sqlstr): # the string is used as an SQL code
    print(str(row[0]), row[1])

cur.close() # closing the connection to the database
```

```python
# Relational Database example; makes a database of all the tracks along with their artists and albums respectively
# from iTunes Library

import xml.etree.ElementTree as ET
import sqlite3

conn = sqlite3.connect('trackdb.sqlite')
cur = conn.cursor()

# Make some fresh tables using executescript()
cur.executescript('''
DROP TABLE IF EXISTS Artist;
DROP TABLE IF EXISTS Album;
DROP TABLE IF EXISTS Track;

CREATE TABLE Artist (
    id  INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
    name    TEXT UNIQUE
);

CREATE TABLE Album (
    id  INTEGER NOT NULL PRIMARY KEY AUTOINCREMENT UNIQUE,
    artist_id  INTEGER,
    title   TEXT UNIQUE
);

CREATE TABLE Track (
    id  INTEGER NOT NULL PRIMARY KEY 
        AUTOINCREMENT UNIQUE,
    title TEXT  UNIQUE,
    album_id  INTEGER,
    len INTEGER, rating INTEGER, count INTEGER
);
''')


fname = input('Enter file name: ')
if ( len(fname) < 1 ) : fname = 'Library.xml'

# <key>Track ID</key><integer>369</integer>
# <key>Name</key><string>Another One Bites The Dust</string>
# <key>Artist</key><string>Queen</string>
def lookup(d, key):
    found = False
    for child in d:
        if found : return child.text
        if child.tag == 'key' and child.text == key :
            found = True
    return None

stuff = ET.parse(fname)
all = stuff.findall('dict/dict/dict')
print('Dict count:', len(all))
for entry in all:
    if ( lookup(entry, 'Track ID') is None ) : continue

    name = lookup(entry, 'Name')
    artist = lookup(entry, 'Artist')
    album = lookup(entry, 'Album')
    count = lookup(entry, 'Play Count')
    rating = lookup(entry, 'Rating')
    length = lookup(entry, 'Total Time')

    if name is None or artist is None or album is None : 
        continue

    print(name, artist, album, count, rating, length)

    cur.execute('''INSERT OR IGNORE INTO Artist (name) 
        VALUES ( ? )''', ( artist, ) )
    cur.execute('SELECT id FROM Artist WHERE name = ? ', (artist, ))
    artist_id = cur.fetchone()[0]

    cur.execute('''INSERT OR IGNORE INTO Album (title, artist_id) 
        VALUES ( ?, ? )''', ( album, artist_id ) )
    cur.execute('SELECT id FROM Album WHERE title = ? ', (album, ))
    album_id = cur.fetchone()[0]

    cur.execute('''INSERT OR REPLACE INTO Track
        (title, album_id, len, rating, count) 
        VALUES ( ?, ?, ?, ?, ? )''', 
        ( name, album_id, length, rating, count ) )

conn.commit()
```
